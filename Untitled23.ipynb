```python
words_from_setn=[]
#function that returns a word in its basic form
def process_sentence(sentence):
    sentence=contractions.fix(sentence)
    words = sentence.split()
    stemmer = PorterStemmer()
    lemmatizer = WordNetLemmatizer()
    processed_words = []
    for word in words:
        word=word.lower()
        word=word.replace(".","")
        word=word.replace(",","")
        word_stem=stemmer.stem(word)
        word_start=lemmatizer.lemmatize(word_stem)
        word_pos_v=lemmatizer.lemmatize(word_start, 'v')
        word_pos_a=lemmatizer.lemmatize(word_pos_v, 'a')
        word_pos_r=lemmatizer.lemmatize(word_pos_a, 'a')
        if "ion" in word_pos_r:
            ind=word_pos_r.index("ion")
            word_ion=word_pos_r[:ind+3]
        else:
            word_ion=word_pos_r
        if word_ion in stop_words:
            word_ion=""
        processed_words.append(word_ion)
        words_from_setn.append(word_ion)
    return ' '.join(processed_words)
```

```python
#Create lists of the length of the opinion and its negative score
ls_all_opinions=[]
len_opinion=[]
opinion_neg_score=[]
sna = SentimentIntensityAnalyzer()
for column_name in opinion_df.columns:
    for index, value in opinion_df[column_name].items():
        ls_all_opinions.append(value)
        sentens=process_sentence(value)
        sent=sentens.split()
        len_opinion.append(len(sent))
        opinion_neg_score.append(sna.polarity_scores(value)['neg']+(sna.polarity_scores(value)['neu']*0.5))
```

```python
#function that accepts a sentence and divides it according to punctuation marks
nlp = spacy.load("en_core_web_sm")
def custom_sentence_splitter_rules(text):
    doc = nlp(text)
    sentences = []
    current_sentence = []
    for token in doc:
        current_sentence.append(token.text)
        if (token.text in (".",",","!", "?", "the")):
            sentences.append(" ".join(current_sentence))
            current_sentence = []
    if current_sentence:
        sentences.append(" ".join(current_sentence))
    return sentences
```

```python
#bag of words
ls_clean = ["clean", "dirty", "musti", "rough", "maintained", "refurbish", "dust", "wipe", "disinfect", "sanitary", "hygiene", "spotless","cleanli"]
ls_internet = ["wifi", "connection", "slow"]
ls_staff = ["staff", "receptionist", "reception", "personnel", "service", "hospitality", "helpful", "friendly", "assistance", "responsive", "host", "hosts", "front desk", "caretaker"]
ls_comfort = ["comfortable", "cozy", "relaxing", "snug", "homely", "pleasant", "peaceful", "spacious", "quiet", "restful", "serene", "comfy", "soothing"]
ls_location = ["location", "nearby", "vicinity", "central", "close", "distant", "proximate", "adjacent", "surrounding", "neighborhood", "area", "site", "surroundings", "vicinity", "locality", "region"]
ls_value_for_money = ["value", "worth", "affordable", "expensive", "cost", "pricey", "budget-friendly", "economical", "reasonable", "overpriced", "cheap", "inexpensive"]
ls_facilities = ["facilities", "amenities", "equipment", "features", "services", "provisions", "offerings", "conveniences", "resources", "installations", "infrastructure", "apparatus", "utilities", "comforts"]

##
cleanliness_words = [element.lower() for element in ls_clean]
hotel_service_words= [element.lower() for element in ls_staff]
internet_words= [element.lower() for element in ls_internet]
hotel_facility_words= [element.lower() for element in ls_facilities]
value_for_money_words= [element.lower() for element in ls_value_for_money]
hotel_location_words= [element.lower() for element in ls_location]
hotel_comfort_words= [element.lower() for element in ls_comfort]
all_words=cleanliness_words+hotel_service_words+internet_words+hotel_facility_words+value_for_money_words+hotel_location_words+hotel_comfort_words
```

```python
#Returns the words found in the category
def get_columns_greater_than_one(index,df):
    columns_greater_than_one = []
    for column_name in df.columns:
        if column_name not in ["sentence", "category"]:
            if df.loc[index, column_name] > 0:  
                columns_greater_than_one.append(column_name)
    return columns_greater_than_one

#cataloging
def Category_senten(column_name,value):
    for word in all_words:
        if (word in column_name) and (value>0) and (word in cleanliness_words):
            return "clean"
        elif (word in column_name) and (word in hotel_service_words) and (value>0):
            return "service"
        elif (word in column_name) and (word in internet_words) and (value>0):
            return "internet"
        elif (word in column_name) and (word in hotel_facility_words) and (value>0):
            return "facility"
        elif (word in column_name) and (word in value_for_money_words) and (value>0):
            return "value_for_money"
        elif (word in column_name) and (word in hotel_location_words) and (value>0):
            return "location"
        elif (word in column_name) and (word in hotel_comfort_words) and (value>0):
            return "comfort"


#def that split sentens whan ther are different categories in one sentence
def split_for_diff(index,greater_than_one,df,ls_name_category):
    indexs_in_sentens=[]
    value=df.loc[index,"sentence"]
    ind_hotel = df_sentens_hotel['sentens'].index[df_sentens_hotel['sentens'] == value][0]
    hotel_name_adding=df_sentens_hotel.loc[ind_hotel,'hotel_name']
    for name in greater_than_one:
        indexs_in_sentens.append(value.index(name))
    
    combined = list(zip(indexs_in_sentens, greater_than_one))
    sorted_combined = sorted(combined)
    sorted_ls = [item[1] for item in sorted_combined]
    greater_than_one=sorted_ls
    combined = list(zip(indexs_in_sentens, ls_name_category))
    sorted_combined = sorted(combined)
    sorted_ls = [item[1] for item in sorted_combined]
    ls_name_category=sorted_ls
    indexs_in_sentens=sorted(indexs_in_sentens)

    for i in range(len(greater_than_one)):
        new_row = [0] * len(df.columns)
        new_sent_1=""
        col_index=df.columns.get_loc(greater_than_one[i])
        if (i<len(greater_than_one)-1) and ls_name_category[i]!=ls_name_category[i+1]:
            if i<(len(greater_than_one)-1):
                new_sent_1=value[:indexs_in_sentens[i+1]]
                value=value[indexs_in_sentens[i+1]:]
                new_row[0]=new_sent_1
                new_row[col_index]=df.loc[index,greater_than_one[i]]
                df_sentens_hotel.loc[len(df_sentens_hotel.index)]=[new_sent_1,hotel_name_adding]
                df.loc[len(df.index)]=new_row
            else:
                new_sent_1=value
                new_row[0]=new_sent_1
                new_row[col_index]=df.loc[index,greater_than_one[i]]
                df_sentens_hotel.loc[len(df_sentens_hotel.index)]=[new_sent_1,hotel_name_adding]
                df.loc[len(df.index)]=new_row
        
        elif i==len(greater_than_one)-1:
            new_sent_1=value
            new_row[0]=new_sent_1
            new_row[col_index]=df.loc[index,greater_than_one[i]]
            df_sentens_hotel.loc[len(df_sentens_hotel.index)]=[new_sent_1,hotel_name_adding]
            df.loc[len(df.index)]=new_row
    return df
```

```python
#Calculation of the weighted score without entering a zero score in the category
df_sort_count=pivot_COUNT.iloc[:,1:].sort_index(axis=1)
df_sort_tf=pivot_TF_sorted.iloc[:,1:].sort_index(axis=1)
df_sort_original=pivot_original_sorted.iloc[:,1:].sort_index(axis=1)
df_sort_original.reset_index(drop=True, inplace=True)
diff_tf=df_sort_count.iloc[:,:].copy()
diff_tf.iloc[:, :] = 0
diff_count=diff_tf.copy()
diff_count_tf=diff_count.copy()
finel_score=diff_count.copy()
for column_name in df_sort_count.columns:
    for index, value in df_sort_count[column_name].items():
        if value!=0 or df_sort_tf.loc[index,column_name]!=0:
            finel_score.loc[index,column_name]=(df_sort_original.loc[index,column_name]+((value*0.5)+(df_sort_tf.loc[index,column_name]*0.5)))/2
        else:
            finel_score.loc[index,column_name]=df_sort_original.loc[index,column_name]+((value*0.5)+(df_sort_tf.loc[index,column_name]*0.5))
        if value!=0:
            diff_count.loc[index,column_name]=(df_sort_original.loc[index,column_name]-value)
            diff_count_tf.loc[index,column_name]=abs(df_sort_tf.loc[index,column_name]-value)
        if df_sort_tf.loc[index,column_name]!=0:
            diff_tf.loc[index,column_name]=(df_sort_original.loc[index,column_name]-df_sort_tf.loc[index,column_name])
```

```python
#Making a hierarchy for the sake of comparison
from scipy.cluster.hierarchy import dendrogram, linkage
features = ['Dist_from_the_beach', 'Dist_from_the_center', 'price', 'opinion_number', 'grade']

Z = linkage(Hotel_df[features], method='average')
plt.figure(figsize=(12, 8))  
dendrogram(
    Z,
    labels=Hotel_df['hotel_name'].values,
    leaf_rotation=90,
    leaf_font_size=12,  
    above_threshold_color='grey',  
)
plt.title('Dendrogram with Average Linkage', fontsize=16)  
plt.xlabel('Hotel', fontsize=14)  
plt.ylabel('Distance', fontsize=14)  
plt.xticks(fontsize=12) 
plt.yticks(fontsize=12)  
plt.grid(True, linestyle='--', alpha=0.6)  
plt.tight_layout()
plt.show()
```

```python
# Finding the shortest distance in the dendrogram
shortest_distance = Z[0, 2]  # The shortest distance is typically the first element in the linkage matrix

# Finding the indices of the pair of hotels with the strongest connection
index_of_hotel1 = int(Z[0, 0])
index_of_hotel2 = int(Z[0, 1])

# Getting the names of the hotels
hotel1_name = Hotel_df['hotel_name'].iloc[index_of_hotel1]
hotel2_name = Hotel_df['hotel_name'].iloc[index_of_hotel2]

# Outputting the pair of hotels with the strongest connection
print(f"The pair of hotels with the strongest connection are '{hotel1_name}' and '{hotel2_name}' with a distance of {shortest_distance}.")
```
